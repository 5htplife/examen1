```python
path = r"/Users/irina/Downloads/GDD_FinalEstimates_01102022/Country-level estimates/"
files = glob.glob(path + "*.csv")

data = pd.DataFrame()
content = []

for filename in files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df = df[df['year'] == 2018]
    df = df.loc[(df['age'] == 999) & (df['female'] == 999)]
    df = df.drop(columns = ['upperci_95', 'lowerci_95', 'serving', 's_lowerci_95', 's_upperci_95'], errors='ignore')
    content.append(df)
```
